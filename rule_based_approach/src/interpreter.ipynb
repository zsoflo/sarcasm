```python
import pandas as pd
import os
import re
from negate import Negator
```

```python
df = pd.read_csv('exp_final_classification_results.csv')
df
```

# Interpretation via Negation

```python
regexp = re.compile(r'(([Hh]+[Aa]+[Tt]+[EeIi]+)|([Ll]+[OoUu]+[Vv]+))\w*\b')
negator = Negator()
df.loc[df['prediction'] == 1, 'negation'] = df["modified_tweet"] #.apply(lambda x: x.lower())
for index, row in df.iterrows():
  if not pd.isna(row['negation']) :
    x = re.search(regexp , row["negation"])
    print(row["negation"])
    noun_phrase = row["negation"][x.span()[1]:]
    without_noun_phrase = row["negation"][:x.span()[1]]
    negated_part = negator.negate_sentence(without_noun_phrase)
    print(negated_part + noun_phrase)
    df.loc[index, "negation"] = negated_part + noun_phrase
```


# Interpreter via sub

```python
hate = re.compile(r'([\b]+[Hh]+[Aa]+[Tt]+)')
love = re.compile(r'([Ll]+[OoUu]+[Vv]+)')

df.loc[df['prediction'] == 1, 'sub'] = df["modified_tweet"]
for index, row in df.iterrows():
  if not pd.isna(row['sub']):
    x = re.search(love , row["sub"])
    res_str = re.sub(love, "hat", row['sub'])
    print(row['sub'])
    print(res_str)
    if not x:
      x = re.search(hate , row["sub"])
      res_str = re.sub(hate, "lov", row['sub'])
    if x:
      before_rest = res_str[:x.span()[0] + 3]
      rest = row["sub"][x.span()[1]:]
      temp = rest.split(' ')
      temp[0] = temp[0].lower()
      rest = ' '.join(temp)
      print(before_rest + rest)
      df.loc[index, "sub"] = before_rest + rest
```
