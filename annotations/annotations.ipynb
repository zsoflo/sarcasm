```python
import pandas as pd
import seaborn as sns
```

```python
florian = pd.read_csv("florian.csv")
ivo = pd.read_csv("ivo.csv")
joely = pd.read_csv("joely.csv")
maiwenn = pd.read_csv("maiwenn.csv")
samba = pd.read_csv("samba.csv")
zsofia = pd.read_csv("zsofia.csv")
```

```python
florian.insert(0, "annotater", "florian")
ivo.insert(0, "annotater", "ivo")
joely.insert(0, "annotater", "joely")
maiwenn.insert(0, "annotater", "maiwenn")
samba.insert(0, "annotater", "samba")
zsofia.insert(0, "annotater", "zsofia")
```

```python
first_100 = [
    ivo.iloc[:100], 
    joely.iloc[:100],
    maiwenn.iloc[:100],
    samba.iloc[:100],
    zsofia.iloc[:100],
    florian.iloc[:100]
]
result = pd.DataFrame(pd.concat(first_100))
```

```python
result["annotation"].fillna(1, inplace = True)
df = pd.DataFrame(result.groupby("annotater").sum(["annotation"]))
```
